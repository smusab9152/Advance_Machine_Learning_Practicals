```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("pima-indians-diabetes.csv", header = None, names = ["no_of_preg","glucose_level","bp","skin_thickness","insulin","BMI","Predigree","Age","Class"])
```

```python
df.head(3)
```

```python
df.shape
```

```python
model = Sequential([
    Dense(units = 10, input_dim = 8, activation = 'relu'),
    Dense(units = 7, activation = 'relu'),
    Dense(units = 1, activation = 'sigmoid')
])
```

```python
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
x = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size = 0.25, random_state = 47)
```

```python
xtrain.shape
```

```python
model.fit(xtrain, ytrain, epochs = 500, batch_size = 20)
```

```python
ypred = model.predict(xtest)
```

```python
ypred
```

```python
class_label = []
for i in range(len(ytest)):
  if (ypred[i] > 0.5):
    class_label.append(1)
  else:
    class_label.append(0)
```

```python
accuracy_score(class_label, ytest)
```
