Design a feed forward neural network that takes 5 features and provides an output.
The FFNN must have a depth of 3.

```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense

nn = Sequential()
```

```python
nn.add(Dense(6, activation ='relu', input_dim = 5))
nn.add(Dense(12, activation ='relu'))
nn.add(Dense(6, activation ='relu'))
nn.add(Dense(1, activation = 'linear')) # for regression - linear activation function, else others
```

```python
nn.compile(optimizer="adam", loss = "mean_absolute_error", metrics = ["accuracy"])
```
