# Collaborative filtering
**Aim: To create a collaborative filtering based model to recommend books**

Note: This is a hybrid approach since titles are in the rows and user ratings are in the columns

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("syedjaferk/book-crossing-dataset")

print("Path to dataset files:", path)
```

```python
import pandas as pd
ratings = pd.read_csv("/root/.cache/kagglehub/datasets/syedjaferk/book-crossing-dataset/versions/1/BX-Book-Ratings.csv", encoding = 'latin-1', delimiter = ';')
ratings.head()
```

```python
books = pd.read_csv("/BX-Books - BX-Books.csv", encoding = 'latin-1', delimiter = ',')
books.head()
```

```python
rating_books = pd.merge(ratings , books, on = 'ISBN')
rating_books.shape
```

```python
ratings_books_sample = rating_books.sample(frac = 0.1, random_state = 47)
ratings_books_sample.shape
```

```python
ratings_books_pivot = ratings_books_sample.pivot_table(index = 'Book-Title', columns = 'User-ID', values = 'Book-Rating').fillna(0)
ratings_books_pivot
```

```python
ratings_books_sample.head(10)['Book-Title']
```

```python
from sklearn.neighbors import NearestNeighbors
nn = NearestNeighbors(metric = 'cosine', algorithm = 'brute', n_neighbors=7, n_jobs = -1)
nn.fit(ratings_books_pivot)
```

```python
indices = nn.kneighbors(ratings_books_pivot.loc[["The Cat Who Went up the Creek"]], 10, return_distance=False)

for index, title in enumerate(ratings_books_pivot.index[indices[0]]):
  print((index+1),".", title)
```
