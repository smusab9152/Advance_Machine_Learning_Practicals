### Gaussian Mixture Model implementation

```python
from sklearn.mixture import GaussianMixture
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Clustering_gmm.csv")
df.head()
```

```python
plt.scatter(df['Weight'], df['Height'])
plt.xlabel("Weight")
plt.ylabel("Height")
plt.show()
```

```python
kmeans = KMeans(n_clusters = 4)
kmeans.fit(df)
```

```python
pred = kmeans.predict(df)
pred
```

```python
df1 = pd.DataFrame(df)
df1['pred'] = pred
df1
```

```python
color = ['red', 'yellow', 'green', 'blue']
for i in range(4):
  data = df1[df1['pred']==i]
  plt.scatter(data['Weight'], data['Height'], c = color[i])
plt.show()
```

```python
df_gmm = pd.read_csv("Clustering_gmm.csv")
gmm = GaussianMixture(n_components=4)
gmm.fit(df_gmm)
```

```python
preds = gmm.predict(df_gmm)
df2 = pd.DataFrame(df_gmm)
df2['preds'] = preds
df2
```

```python
color = ['red', 'yellow', 'green', 'blue']
for i in range(4):
  data = df2[df2['preds']==i]
  plt.scatter(data['Weight'], data['Height'], c = color[i])
plt.show()
```
