```python
import pandas as pd
from keras.layers import Dense
from keras.models import Sequential
```

```python
from sklearn.datasets import load_iris
data = load_iris()
x = data.data
y = data.target
```

```python
x
```

```python
y
```

```python
seq = Sequential()
```

```python
seq.add(Dense(8,activation='relu',input_dim = 4))
seq.add(Dense(6,activation='relu'))
seq.add(Dense(3,activation='softmax'))
```

```python
from sklearn.model_selection import train_test_split
```

```python
from keras.utils import to_categorical
encded_y = to_categorical(y)
encded_y
```

```python
xtrain,xtest , ytrain,ytest = train_test_split(x , encded_y , test_size=0.2 , random_state= 4)
```

```python
seq.compile(loss = 'categorical_crossentropy', optimizer = 'adam',metrics = ['accuracy'])
```

```python
seq.fit(xtrain , ytrain , epochs= 100 ,batch_size = 20)
```

```python
pred = seq.predict(xtest)
```

```python
for i in range(0,30,3):
  print(f'{pred[i] }       {ytest[i]}')
```

```python
import numpy as np
pred_class =[]
for i in range(len(pred)):
  pred_class.append(np.argmax(pred[i] , axis =1))

```
